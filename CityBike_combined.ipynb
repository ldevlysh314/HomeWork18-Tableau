```python
import pandas as pd
```

```python
# Data files to be combined 
files23 = ['data/JC-202310-citibike-tripdata.csv', 'data/JC-202311-citibike-tripdata.csv', 'data/JC-202312-citibike-tripdata.csv']

# Empty list to store each dataframe
df23_list = []

# Loop through each file in the list
for file in files23:
    # Read the file into a dataframe
    df23 = pd.read_csv(file)

    # Append the dataframe to the list
    df23_list.append(df23)

# Concatenate all dataframes in the list into one dataframe
df23 = pd.concat(df23_list)

# Show dataframe
df23
```

```python
# Save the dataframe to a CSV file
df23.to_csv('data/citibike23.csv', index=False)
```
